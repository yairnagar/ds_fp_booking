```python
# get the number of pages in current city.
pages = int(driver.find_elements(By.CLASS_NAME, "f9c5690c58")[-1].text)
```

```python
def getURLS(driver):
    current_page_urls = []
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    for tab in soup.findAll(attrs={'class': "d20f4628d0"}):
        current_page_urls.append(tab.h3.find("a")['href'])
    
    return current_page_urls
```

```python
for i in range(pages):
    main_page = driver.current_url
    hotelsUrl = getURLS(driver)
    print("moving page ..")
    print(f"page number : {i+1}")
    for j in range(1, len(hotelsUrl)):
        print(f"hotel number {j} in this page")
        driver.get(hotelsUrl[j])
        time.sleep(2)
        
        # Parsing the page
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        
        try:
            # all score components:
            scores_component = soup.findAll('div', {"class": "bui-spacer--larger"})[1]
            scores_component = scores_component.findAll('div', {"class": "b8eef6afe1"})
            staff_score.append(scores_component[0].get_text())
            print(f"Staff : {scores_component[0].get_text()}")
            facilities_score.append(scores_component[1].get_text())
            print(f"Facilities : {scores_component[1].get_text()}")
            cleanliness_score.append(scores_component[2].get_text())
            print(f"Cleanliness : {scores_component[2].get_text()}")
            comfort_score.append(scores_component[3].get_text())
            print(f"Comfort : {scores_component[3].get_text()}")
            value_for_money_score.append(scores_component[4].get_text())
            print(f"Value for money : {scores_component[4].get_text()}")
            location_score.append(scores_component[5].get_text())
            print(f"Location : {scores_component[5].get_text()}")
            
            # Wifi score not always avaliable in scores components
            if (len(scores_component) == 7):
                free_wifi_score.append(scores_component[6].get_text())
                print(f"Free Wifi : {scores_component[6].get_text()}")
            else:
                free_wifi_score.append(np.nan)
            
            # get number of stars
            num_stars = driver.find_elements(By.CLASS_NAME, "adc357e4f1")
            if num_stars:
                stars.append(len(num_stars))
                print(f"Stars : {len(num_stars)}")
            else:
                stars.append(np.nan)
            
            # get hotel's name
            hotelName = driver.find_element(By.CLASS_NAME, "d2fee87262")
            if hotelName.text:
                names.append(hotelName.text)
                print(f"Name : {hotelName.text}")
            else:
                names.append(np.nan)
                
                    
            # get hotel's price
            hotelPrice = driver.find_element(By.CLASS_NAME, "bui-price-display__value")
            if hotelPrice:
                prices.append(hotelPrice.text.split()[1])
                print(f"Price : {hotelPrice.text.split()[1]}")
            else:
                prices.append(np.nan)
            
            # get hotel's rating
            hotelRating = driver.find_element(By.CLASS_NAME, "d10a6220b4")
            if hotelRating:
                ratings.append(hotelRating.text)
                print(f"Rating : {hotelRating.text}")
            else:
                ratings.append(np.nan)
            
            # get number of reviewes
            ratingSumStr = driver.find_element(By.CLASS_NAME, "db63693c62")
            if ratingSumStr:
                reviews.append(ratingSumStr.text.split()[0])
                print(f"Reviews : {ratingSumStr.text.split()[0]}")
            else:
                reviews.append(np.nan)
            


            try:
                
                distances_component = driver.find_elements(By.CLASS_NAME, "d31796cb42")

                # get closest attraction distance
                value = [i.text.split('\n')[2].split()[0] for i in distances_component if
                            i.text.split('\n')[0] == "Top attractions"]
                if len(value) != 0:
                    closest_attraction.append(value)
                    print(f"Closest attraction :  {value}")
                else:
                    closest_attraction.append(np.nan)
                
                
                # get closest transportation distance
                value = [i.text.split('\n')[2].split()[0] for i in distances_component if
                        i.text.split('\n')[0] == "Public transit"]
                if len(value) != 0:
                    closest_transportation.append(value)
                    print(f"Closest transportation :  {value}")
                else:
                    closest_transportation.append(np.nan)

                # get closest airport
                value = [i.text.split('\n')[2].split()[0] for i in distances_component if 
                        i.text.split('\n')[0] == "Closest Airports"]
                if len(value) != 0:
                    closest_airport.append(value)
                    print(f"Closest airport :  {value}")
                else:
                    closest_airport.append(np.nan)


            except:
                closest_attraction.append(np.nan)
                closest_transportation.append(np.nan)
                closest_airport.append(np.nan)


            #get number of spoken languages
            try:
                hotel_facilities_container = driver.find_elements(By.CLASS_NAME, "bui-spacer--large")
                value = [len(i.text.split('\n')[1:]) for i in hotel_facilities_container if
                        i.text.split('\n')[0] == "Languages Spoken"]
                if len(value) != 0:
                    languages_spoken.extend(value)
                    print(f"Number of languages: {value}")
                else:
                    languages_spoken.append(np.nan)
            except:
                    languages_spoken.append(np.nan)


            
            ## get if pets allowed
            try:
                pets_div = driver.find_elements(By.CLASS_NAME, "description--house-rule")[0]
                pets_phrase = pets_div.text.split('\n')[1].split('.')[0]
                if pets_phrase == 'Pets are allowed on request' or pets_phrase == 'Pets are allowed' :
                    print("Is pets allowed : True")
                    pets_allowed.append(True)
                elif pets_phrase == 'Pets are not allowed' :
                    print("Is pets allowed : False")
                    is_pets_allowed.append(False)
                else:
                    pets_allowed.append(np.nan)

            except:
                pets_allowed.append(np.nan)


            ## get num of credit cards logos
            try:
                cards_accepted = driver.find_element(By.CLASS_NAME, "payment_methods_overall")
                value = len(cards_accepted.find_elements(By.TAG_NAME, "img"))
                if value != 0:
                    payment_methods.append(value)
                    print(f"Number of payments methods : {value}")
                else:
                    payment_methods.append(np.nan)

            except:
                payment_methods.append(np.nan)
        
        except:
            print("cant find something")



    # after crawling all hotesl in current page, moving to the next page if is exist
    driver.get(main_page)
    if i == 0:
            driver.find_elements(By.CLASS_NAME, "e1b7cfea84")[0].click()
            print(driver.find_elements(By.CLASS_NAME, "e1b7cfea84")[0].text)
            time.sleep(2)
    elif i != pages - 1 :
        driver.find_elements(By.CLASS_NAME, "e1b7cfea84")[1].click()
        print(driver.find_elements(By.CLASS_NAME, "e1b7cfea84")[1].text)
        time.sleep(2)
```
