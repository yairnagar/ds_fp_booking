```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV files from List
hotels = pd.concat(map(pd.read_csv, ['Clean_Atlanta.csv', 'Clean_Boston.csv', 'Clean_Chicago.csv', 'Clean_Dallas.csv', 'Clean_Houston.csv'
, 'Clean_LasVegas.csv', 'Clean_Miami.csv', 'Clean_NewOrleans.csv', 'Clean_NewYork.csv', 'Clean_Orlando.csv'
, 'Clean_Phoenix.csv', 'Clean_SanDiego.csv', 'Clean_Washington.csv']), ignore_index=True)
hotels.head(60)
hotels.info()
```

```python
hotels['City'].value_counts().plot(kind='bar', figsize=(10, 10))
plt.title('Number of hotels in each City')
```

```python
hotels['City'].value_counts().plot(kind='pie',autopct="%.2f%%", figsize= (20,10))
plt.title('Number of hotels in each City')
```

```python
T = sns.catplot(
    x='City', 
    data = hotels,
    kind='count',
    hue='Stars',
    height=5, 
    aspect=4,
    legend=False,
    ).set_axis_labels('City', 'Number of Hotels')
T.ax.legend(labels=['1-Star','2-Stars','3-Stars','4-Stars','5-Stars'])
```

```python
hotels.plot(x='Reviews', y= 'Rating', kind='scatter',title= 'Rating score : Number of reviews')
```

```python
hotels.plot(x='Location', y= 'Rating', kind='scatter',title= 'Rating score : Location score')
```

```python
hotels.plot(x='Rating', y= 'Price', kind='scatter',title= 'Price for one night : Rating score')
```

```python
hotels.plot(x='Rating', y= 'Value for Money', kind='scatter',title= 'Rating score :Value for money score ')
```

```python
hotels.plot(x='Location', y= 'Price', kind='scatter',title= 'Price for one night : Location score')
```

```python
hotels.plot(x='staff', y= 'Price', kind='scatter',title= 'Price fro one night : Staff score')
```

```python
prices = [hotels[hotels['Allowing pets']==True].Price.median(),
          hotels[hotels['Allowing pets']==False].Price.median()]
plt.figure()
plt.subplot(111)
plt.axis('on')
plt.title('Price for one : Is Allowing Pets')
plt.bar(['Allowing pets','Not allowing pets'], prices)
```

```python
cleanScore = [hotels[hotels['Allowing pets']==True].Cleanliness.median(),
              hotels[hotels['Allowing pets']==False].Cleanliness.median()]
print(cleanScore)
plt.figure()
plt.subplot(111)
plt.axis('on')
plt.title('Cleanliness score : Is Allowing Pets')
plt.bar(['Allowing pets','Not allowing pets'], cleanScore)
```
